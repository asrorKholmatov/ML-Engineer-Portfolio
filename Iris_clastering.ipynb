```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


sns.scatterplot(
    data=iris_df,
    x='sepal_length',
    y='petal_length',
    hue='species',
    palette='Set2',
    s=100,
    edgecolor='black',
    linewidth=0.5
)


plt.title('Sepal vs Petal Length by Species', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species', loc='upper left')

plt.tight_layout()
plt.show()
```

```python
numeric_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

X = iris_df[numeric_cols]
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
```

```python
model.cluster_centers_
```

```python
preds = model.predict(X)
preds
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


sns.scatterplot(
    data=iris_df,
    x='sepal_length',
    y='petal_length',
    hue=preds,
    palette='Set2',
    s=100,
    edgecolor='black',
    linewidth=0.5
)


plt.title('Sepal vs Petal Length by Preds', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Preds', loc='upper left')

centers_x, centers_y = model.cluster_centers_[:,0], model.cluster_centers_[:,2]
plt.plot(centers_x, centers_y, 'xb', markerfacecolor='yellow', markeredgecolor='red',  markeredgewidth=3, markersize=10 )



plt.tight_layout()
plt.show()
```

```python
model.inertia_
```

```python
model = KMeans(n_clusters=6, random_state=42).fit(X)
```

```python
preds = model.predict(X)
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


sns.scatterplot(
    data=iris_df,
    x='sepal_length',
    y='petal_length',
    hue=preds,
    palette='Set2',
    s=100,
    edgecolor='black',
    linewidth=0.5
)


plt.title('Sepal vs Petal Length by Preds', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Preds', loc='upper left')

centers_x, centers_y = model.cluster_centers_[:,0], model.cluster_centers_[:,2]
plt.plot(centers_x, centers_y, 'xb', markerfacecolor='yellow', markeredgecolor='red',  markeredgewidth=3, markersize=10 )



plt.tight_layout()
plt.show()
```

```python
options = range(2,11)
inertias = []

for n_clusters in options:
    model = KMeans(n_clusters=n_clusters, random_state=42).fit(X)
    inertias.append(model.inertia_)

sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(8, 5))


plt.plot(options, inertias, '-o', color='teal', linewidth=2, markersize=8, markerfacecolor='orange')


plt.title("Number of Clusters vs Inertia", fontsize=16, weight='bold')
plt.xlabel("Number of Clusters (k)", fontsize=12)
plt.ylabel("Inertia", fontsize=12)


plt.xticks(options)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps=1.5, min_samples=4)
model.fit(X)
```

```python
model.labels_
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


sns.scatterplot(
    data=X,
    x='sepal_length',
    y='petal_length',
    hue=model.labels_,
    palette='Set2',
    s=100,
    edgecolor='black',
    linewidth=0.5
)


plt.title('Sepal vs Petal Length by Preds', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Preds', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = model.fit_predict(X)
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


plt.scatter(X.iloc[:, 0],
            X.iloc[:, 1],
            c=labels,
            cmap='viridis')

plt.title('Sepal vs Petal Length by Preds', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Preds', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(iris_df[numeric_cols])
```

```python
X_pca = pca.transform(iris_df[numeric_cols])
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 6))


sns.scatterplot(
    x=X_pca[:,0],
    y=X_pca[:,1],
    hue=iris_df['species'],
    palette='Set2',
    s=100,
    edgecolor='black',
    linewidth=0.5
)


plt.title('Sepal vs Petal Length by Preds', fontsize=16, weight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Preds', loc='upper left', frameon=True)
plt.tight_layout()
plt.show()
```
